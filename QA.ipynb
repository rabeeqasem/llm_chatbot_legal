```python
import json

with open('text_files\law_dict.json', encoding='utf-8') as f:
    law_dict = json.load(f)
```

```python
key_list=list(law_dict.keys())
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages= messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
```

```python
# create a function that take a string and return a promt
def get_promt(text):
    promt=f"""

    اكتب لي 5 اسئلة واجوبتها من الفقرة التالية

    --------------------------------------
    الفقرة:

    {text}
    """
    return promt
```

```python
import time
```

```python
import time

total_articles = len(key_list)
for i, article in enumerate(key_list):
    prompt = get_promt(law_dict[article])
    result = get_completion(prompt)
    
    # Write the result to the output.txt file without overwriting, appending at the end
    with open('output.txt', 'a', encoding='utf-8') as f:
        f.write(result)
        f.write("\n\n")
    
    print(article)
    print("\n")
    # Display progress
    progress = (i + 1) / total_articles * 100
    print(f"Progress: {progress:.2f}%")
    
    # Wait for 10 seconds
    time.sleep(10)
```
